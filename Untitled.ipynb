```python
#Import needed libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

df = pd.read_csv('monster_data.csv')

df.groupby('challenge')['name'].count()
```

Determine AC by Level

```python
def bar(column, color = 'blue'):
    bar_df = df.groupby('challenge')[column].mean()
    bar_df.plot(kind='bar', color = color)
    plt.title('Average ' + column + ' by Challenge')
    plt.ylabel('Average ' + column)
    plt.xlabel('Challenge')
    plt.legend()

bar('hp', color = 'red')
bar('ac')
bar('total_save_bonus', color='green')

```
